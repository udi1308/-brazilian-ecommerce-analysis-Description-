## IMPORTING OF LIBRARY

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## IMPORTING OF DATASETS

```python
customers = pd.read_csv('olist_customers_dataset.csv')
orders = pd.read_csv('olist_orders_dataset.csv')
items = pd.read_csv('olist_order_items_dataset.csv')
payments = pd.read_csv('olist_order_payments_dataset.csv')
reviews = pd.read_csv('olist_order_reviews_dataset.csv')
products = pd.read_csv('olist_products_dataset.csv')
```

```python
print(customers.head(5))
print('--------------------')

print(orders.head(5))
print('--------------------')

print(items.head(5))
print('--------------------')

print(payments.head(5))
print('--------------------')

print(reviews.head(5))
print('--------------------')


print(products.head(5))
```

## MERGE OF DATASET 

```python
df=orders.merge(customers,on='customer_id',how='left')
df=df.merge(items,on='order_id',how='left')
df=df.merge(payments,on='order_id',how='left')
df=df.merge(products,on='product_id',how='left')
df=df.merge(reviews,on='order_id',how='left')
df.head()
```

## CLEAN AND TRANSFORM

```python
# convert dates
df['order_purchase_timestamp']=pd.to_datetime(df['order_purchase_timestamp'])
df['order_delivered_customer_date'] = pd.to_datetime(df['order_delivered_customer_date'])
```

```python
# Create TotalPrice
df['Total_Price'] = df['price'] + df['freight_value']
```

```python
# Remove Nulls
df= df.dropna(subset=['order_purchase_timestamp'])
```

```python
df.to_csv('Final_Datase.csv',index=False)
```

```python
df['order_month'] = df['order_purchase_timestamp'].dt.to_period('M')
df['order_year'] = df['order_purchase_timestamp'].dt.year
```

## EDA

```python
# Monthly Revenue
monthly_revenue = df.groupby('order_month')['Total_Price'].sum()
monthly_revenue.head()
```

```python
# Top States
states_sales = df.groupby('customer_state')['Total_Price'].sum().sort_values(ascending=False)
states_sales.head()
```

```python
# Top Products
top_products = df.groupby('product_id')['Total_Price'].sum().sort_values(ascending=False)
top_products.head()
```

```python
# Repeat Customer
repeat_customer = df.groupby('customer_unique_id')['order_id'].nunique()
repeat_customer = repeat_customer[repeat_customer > 1]
repeat_customer.head()
```
